```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
print(tf.__version__)
```

```python
imdb = keras.datasets.imdb

(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = 10000)
```

```python
# A dictionary mapping words to an integer index
word_index = imdb.get_word_index()

# The first indices are reserved
word_index = {k:(v+3) for k,v in word_index.items()} 
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2  # unknown
word_index["<UNUSED>"] = 3

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])
```

```python
decode_review(train_data[3])
```

```python
len(word_index)
```

```python
# Prepare the data
# Create an integer tensor of shape max_length * numn_reviews. 
#Use an embedding layer capable of handling this shape as the first alyer in our network

train_data = keras.preprocessing.sequence.pad_sequences(train_data,value=word_index["<PAD>"], padding='post', maxlen=256)

test_data = keras.preprocessing.sequence.pad_sequences(test_data,value=word_index["<PAD>"], padding='post', maxlen=256)
```

```python
len(train_data[0]), len(train_data[3])
```

```python
print(train_data[3])
```

```python
# BUild a model 
# How many layers to use in the model?
# How many hidden units to use fo each layer?

# input layer shape is the vocabulary count used for the movie reviews (10,000 words)

vocab_size = 10000

model = keras.Sequential()
model.add(keras.layers.Embedding(input_dim=vocab_size,output_dim=16))
model.add(keras.layers.GlobalAveragePooling1D())
model.add(keras.layers.Dense(units=16, activation=tf.nn.relu))
model.add(keras.layers.Dense(units=1, activation=tf.nn.sigmoid))

model.summary()
```

```python
# Model Compile

model.compile(optimizer=tf.train.AdamOptimizer(),
             loss='binary_crossentropy',
             metrics = ['accuracy'])
```

```python
# create a validation set

x_val = train_data[:10000]
partial_x_train = train_data[10000:]

y_val = train_labels[:10000]
partial_y_train = train_labels[10000:]
```

```python
#Train the Model
history = model.fit(x=partial_x_train,y=partial_y_train,epochs=40,batch_size=512, validation_data=(x_val,y_val), verbose=1)
```

```python
# Evaluvate the model

results = model.evaluate(test_data, test_labels)

print(results)
```

```python
# create a graph graph of accuracy and loss over time

history_dict = history.history
history_dict.keys()
```
